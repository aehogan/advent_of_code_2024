```python
import numpy as np
import re
```

```python
with open("input.txt") as f:
    lines = f.readlines()

machines = []
machine = None

for line in lines:
    line = line.strip()
    if machine is None:
        machine = {"A": [], "B": [], "Prize": []}
    if len(line) == 0:
        machines.append(machine)
        machine = None
    elif line[:5] == "Prize":
        pattern = r"X=([0-9]*), Y=([0-9]*)"
        result = re.search(pattern, line)
        machine["Prize"] = np.array([int(result.group(1)), int(result.group(2))])
        machine["Prize"] += 10000000000000
    elif line[:8] == "Button A":
        pattern = r"X\+([0-9]*), Y\+([0-9]*)"
        result = re.search(pattern, line)
        machine["A"] = np.array([int(result.group(1)), int(result.group(2))])
    elif line[:8]:
        pattern = r"X\+([0-9]*), Y\+([0-9]*)"
        result = re.search(pattern, line)
        machine["B"] = np.array([int(result.group(1)), int(result.group(2))])

if machine is not None:
    machines.append(machine)
    machine = None

machines
```

```python
total = 0
for machine in machines:
    b = 1/(machine['B'][0]-machine['B'][1]*machine['A'][0]/machine['A'][1])*(machine['Prize'][0]-machine['Prize'][1]*machine['A'][0]/machine['A'][1])
    a = (machine['Prize'][1]-b*machine['B'][1])/machine['A'][1]
    rounded_ab = np.rint(np.array([a, b])).astype(int)
    if np.all(rounded_ab[0]*machine['A'] + rounded_ab[1]*machine['B'] == machine['Prize']):
        total += 3*rounded_ab[0] + rounded_ab[1]
total
```
