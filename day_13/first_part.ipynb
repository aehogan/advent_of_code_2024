```python
import numpy as np
import re
```

```python
with open("input.txt") as f:
    lines = f.readlines()

machines = []
machine = None

for line in lines:
    line = line.strip()
    if machine is None:
        machine = {"A": [], "B": [], "Prize": []}
    if len(line) == 0:
        machines.append(machine)
        machine = None
    elif line[:5] == "Prize":
        pattern = r"X=([0-9]*), Y=([0-9]*)"
        result = re.search(pattern, line)
        machine["Prize"] = np.array([int(result.group(1)), int(result.group(2))])
    elif line[:8] == "Button A":
        pattern = r"X\+([0-9]*), Y\+([0-9]*)"
        result = re.search(pattern, line)
        machine["A"] = np.array([int(result.group(1)), int(result.group(2))])
    elif line[:8]:
        pattern = r"X\+([0-9]*), Y\+([0-9]*)"
        result = re.search(pattern, line)
        machine["B"] = np.array([int(result.group(1)), int(result.group(2))])

if machine is not None:
    machines.append(machine)
    machine = None

machines
```

```python
total = 0
for machine in machines:
    scores = []
    for a_presses in range(100):
        target = machine['Prize']-a_presses*machine['A']
        if np.all(target%machine['B'] == 0):
            if (target//machine['B'])[0] == (target//machine['B'])[1]:
                b_presses = (target//machine['B'])[0]
                score = 3*a_presses + b_presses
                scores.append(score)
    if len(scores) != 0:
        total += np.sort(scores)[0]
total
```
