```python
import numpy as np
import itertools
```

```python
with open('input.txt') as f:
    data = f.readlines()

heights = []
for line in data:
    height = []
    for char in line.strip():
        height.append(int(char))
    heights.append(height)
heights = np.array(heights)
print(heights.shape)
heights
```

```python
trailheads = []
for y, row in enumerate(heights):
    for x, height in enumerate(row):
        if height == 0:
            trailheads.append(np.array([x,y]))
trailheads
```

```python
shifts = (np.array([1, 0]), np.array([-1, 0]), np.array([0, 1]), np.array([0, -1]))
def get_nearby_cells(start_pos):
    nearby = []
    for shift in shifts:
        pos = start_pos + shift
        if np.all(pos >= 0) and np.all(pos < heights.shape):
            nearby.append(pos)
    return nearby
```

```python
def is_valid_path(path):
    #print(path)
    for idx, pos in enumerate(path):
        #print(idx, pos)
        if idx == 9:
            return True
        if not any((path[idx+1] == x).all() for x in get_nearby_cells(pos)):
            return False
```

```python
total = 0
for trailhead in trailheads:
    map = -np.ones_like(heights)
    map[trailhead[1], trailhead[0]] = 0
    current_pos = [trailhead]
    heights_dict = {0: current_pos}
    #print(f'start {trailhead}')
    for current_height in range(1, 10):
        new_pos = []
        for pos in current_pos:
            nearby_cells = get_nearby_cells(pos)
            for cell in nearby_cells:
                if map[cell[1], cell[0]] == -1 and heights[cell[1], cell[0]] == current_height:
                    map[cell[1], cell[0]] = current_height
                    new_pos.append(cell)
        current_pos = new_pos
        heights_dict[current_height] = current_pos
        #print(current_height, current_pos)
    #print(map >= 0)
    #print(heights == 9)
    #print(heights_dict)
    for path in itertools.product(heights_dict[0], heights_dict[1], heights_dict[2], heights_dict[3], heights_dict[4], heights_dict[5], heights_dict[6], heights_dict[7], heights_dict[8], heights_dict[9]):
        if is_valid_path(path):
            #print(path)
            total += 1
    #total += np.count_nonzero(np.logical_and(map >= 0, heights == 9))
total
```
