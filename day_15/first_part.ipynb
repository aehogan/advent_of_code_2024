```python
import numpy as np
import itertools
```

```python
dir_to_number = {'^': 0, '>': 1, 'v': 2, '<': 3}
map_element_to_number = {'.': 0, '@': 0, '#': 1, 'O': 0}
map = []
bots = []
barrels = []

with open('input.txt') as f:
    for y, line in enumerate(itertools.takewhile(lambda x: x != '\n', f)):
        row = []
        for x, char in enumerate(line.strip()):
            row.append(map_element_to_number[char])
            if char == '@':
                bots.append(np.array([x, y]))
            elif char == 'O':
                barrels.append(np.array([x, y]))
        map.append(row)
    map = np.array(map)
    
    instructions = []
    for line in f:
        for char in line.strip():
            instructions.append(dir_to_number[char])

print(bots)
print(barrels)
print(map)
print(instructions)
```

```python
def draw():
    for y, row in enumerate(map):
        for x, ele in enumerate(row):
            coord = np.array([x, y])
            found_bot = False
            for bot in bots:
                if np.all(bot == coord):
                    print('@', end='')
                    found_bot = True
            if found_bot:
                continue
            elif np.any([np.all(coord == barrel) for barrel in barrels]):
                print('O', end='')
            else:
                if ele == 0:
                    print('.', end='')
                elif ele == 1:
                    print('#', end='')
        print()

draw()
```

```python
def instruction_to_array(instruction):
    if instruction == 0:
        return np.array((0, -1))
    elif instruction == 1:
        return np.array((1, 0))
    elif instruction == 2:
        return np.array((0, 1))
    elif instruction == 3:
        return np.array((-1, 0))
    else:
        raise ValueError
```

```python
def look_ahead(position, direction):
    looking_good = True
    if map[position[1], position[0]] == 1:
        looking_good = False
    found_barrel = False
    for idx, barrel in enumerate(barrels):
        if np.all(position == barrel):
            found_barrel = True
            break
    if found_barrel:
        if look_ahead(position + direction, direction):
            barrels[idx] = position + direction
        else:
            looking_good = False
    #print(position, direction, looking_good, found_barrel)
    return looking_good
```

```python
for instruction in instructions:
    #draw()
    for idx, bot in enumerate(bots):
        direction = instruction_to_array(instruction)
        look_here = direction + bot
        if look_ahead(look_here, direction):
            bots[idx] = look_here
```

```python
total = 0
for barrel in barrels:
    total += barrel[1] * 100 + barrel[0]
total
```
