```python
import numpy as np
import itertools
```

```python
dir_to_number = {'^': 0, '>': 1, 'v': 2, '<': 3}
map_element_to_number = {'.': 0, '@': 0, '#': 1, 'O': 0}
map = []
bots = []
barrels = []

with open('input_small.txt') as f:
    for y, line in enumerate(itertools.takewhile(lambda x: x != '\n', f)):
        row = []
        for x, char in enumerate(line.strip()):
            row.append(map_element_to_number[char])
            if char == '@':
                bots.append(np.array([x, y]))
            elif char == 'O':
                barrels.append(np.array([x, y]))
        map.append(row)
    map = np.array(map)
    
    instructions = []
    for line in f:
        for char in line.strip():
            instructions.append(dir_to_number[char])

print(bots)
print(barrels)
print(map)
print(instructions)
```

```python
def draw():
    for y, row in enumerate(map):
        for x, ele in enumerate(row):
            coord = np.array([x, y])
            if np.all(bot == coord):
                print('@', end='')
            elif np.any([np.all(coord == barrel) for barrel in barrels]):
                print('O', end='')
            else:
                if ele == 0:
                    print('.', end='')
                elif ele == 1:
                    print('#', end='')
        print()

draw()
```

```python
def instruction_to_array(instruction):
    if instruction == 0:
        return np.array((0, 1))
    elif instruction == 1:
        return np.array((1, 0))
    elif instruction == 2:
        return np.array((0, -1))
    elif instruction == 3:
        return np.array((-1, 0))
    else:
        raise ValueError
```

```python
def look_ahead(position, direction):
    looking_good = True
    if map[position[0], position[1]] == 1:
        looking_good = False
    return looking_good
```

```python
for instruction in instructions:
    draw()
    for bot in bots:
        direction = instruction_to_array(instruction)
        look_here = direction + bot
        if look_ahead(look_here, direction):
            bot = look_here
```
