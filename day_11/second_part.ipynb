```python
import numpy as np
```

```python
with open("input.txt") as f:
    input_data = f.readlines()[0].strip().split()
input_data = np.array([int(x) for x in input_data], dtype=int)
input_data
```

```python
total = 0
#first_iters = 13
#second_iters = 12
first_iters = 38
second_iters = 37
val_to_eventual_length = {}

for a_idx, starting_point in enumerate(input_data):
    print(f"---{a_idx}---")
    data = [starting_point]
    array = np.empty(1000000000, dtype=int)
    n_stones = len(data)
    for idx in range(n_stones):
        array[idx] = data[idx]
    for _ in range(first_iters):
        for idx in range(n_stones):
            number = array[idx]
            str_number = str(number)
            len_number = len(str_number)
            if number == 0:
                array[idx] = 1
            elif len_number % 2 == 0:
                array[idx] = int(str_number[:len_number//2])
                array[n_stones] = int(str_number[len_number//2:])
                n_stones += 1
            else:
                array[idx] *= 2024
        #print(_, n_stones)
    
    first_array = np.copy(array[:n_stones])
    distinct_vals = np.sort(np.array(list(set(list(array[:n_stones])))))
    
    #print(distinct_vals)
    #print(len(distinct_vals))

    print("first part done")
    
    for val_idx, val in enumerate(distinct_vals):
        if val in val_to_eventual_length.keys():
            continue
        n_stones = 1
        array[0] = val
        for _ in range(second_iters):
            for idx in range(n_stones):
                number = array[idx]
                str_number = str(number)
                len_number = len(str_number)
                if number == 0:
                    array[idx] = 1
                elif len_number % 2 == 0:
                    array[idx] = int(str_number[:len_number//2])
                    array[n_stones] = int(str_number[len_number//2:])
                    n_stones += 1
                else:
                    array[idx] *= 2024
        val_to_eventual_length[val] = n_stones
        #print(val_idx, val, n_stones)
    
    
    for x in first_array:
        total += val_to_eventual_length[x]
    print("second part done")
print("total", total)
```
