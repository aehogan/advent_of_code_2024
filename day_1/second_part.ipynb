```python
import numpy as np
```

```python
input = np.loadtxt("input", dtype=int)
input
```

```python
left = np.sort(input[:,0])
right = np.sort(input[:,1])
print(left)
print(right)
```

```python
right_values, right_counts = np.unique(right, return_counts=True)
```

```python
score = 0
for x in left:
    if x in right_values:
        count = right_counts[np.where(right_values==x)[0][0]]
        score += x*count
score
```
