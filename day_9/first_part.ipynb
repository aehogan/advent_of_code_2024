```python
import numpy as np
```

```python
with open('input.txt') as f:
    input = f.readlines()[0]
input
```

```python
total_size = 0
for char in input:
    total_size += int(char)
total_size
```

```python
data = -np.ones(total_size, dtype=int)
current_file = 0
current_pointer = 0
for char_idx, char in enumerate(input):
    size = int(char)
    if char_idx % 2 == 0:
        data[current_pointer:current_pointer+size] = current_file
        current_file += 1
    current_pointer += size
data
```

```python
left_ptr = 0
right_ptr = total_size - 1
while True:
    while data[left_ptr] != -1:
        left_ptr += 1
    while data[right_ptr] == -1:
        right_ptr -= 1
    if left_ptr > right_ptr:
        break
    print(left_ptr, right_ptr, data[left_ptr], data[right_ptr])
    data[left_ptr] = data[right_ptr]
    data[right_ptr] = -1
data
```

```python
print(left_ptr)
print(right_ptr)
```

```python
total = 0
for idx, x in enumerate(data):
    if x == -1:
        continue
    total += idx*x
total
```
