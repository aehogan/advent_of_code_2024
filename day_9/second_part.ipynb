```python
import numpy as np
```

```python
with open('input.txt') as f:
    input = f.readlines()[0]
input
```

```python
total_size = 0
for char in input:
    total_size += int(char)
total_size
```

```python
data = -np.ones(total_size, dtype=int)
files = []
current_file = 0
current_pointer = 0
for char_idx, char in enumerate(input):
    size = int(char)
    if char_idx % 2 == 0:
        data[current_pointer:current_pointer+size] = current_file
        files.append((current_file, size, current_pointer))
        current_file += 1
    current_pointer += size
files.reverse()
print(data)
print(files)
```

```python
def find_free_space(target_size):
    free_space_start_idx = 0
    free_space_current_idx = 0
    while free_space_current_idx < total_size:
        free_space_size = 0
        while data[free_space_start_idx] != -1:
            free_space_start_idx += 1
        free_space_current_idx = free_space_start_idx
        while data[free_space_current_idx] == -1:
            free_space_size += 1
            free_space_current_idx += 1
            if free_space_current_idx >= total_size:
                break
        if free_space_size >= target_size:
            return free_space_start_idx
        else:
            free_space_start_idx += free_space_size
    return False
```

```python
#print(data)
for file_id, size, ptr in files:
    move_idx = find_free_space(size)
    #print(file_id, size, move_idx)
    if move_idx is False or move_idx >= ptr:
        continue
    data[data==file_id] = -1
    data[move_idx:move_idx+size] = file_id
    #print(data)
```

```python
total = 0
for idx, x in enumerate(data):
    if x == -1:
        continue
    total += idx*x
total
```
